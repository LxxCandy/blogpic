```python
from openai import OpenAI

client = OpenAI(
    base_url='https://ark.cn-beijing.volces.com/api/v3/',
    api_key='c7889011-306b-42ef-8d7e-bd86d08bb985',
)
model = 'ep-20251023231522-49jmk'

prompt = """
你正在协助我分析专利文本以识别核心技术要素。专利名称和摘要将作为用户提示提供。你的目标是从中提取具有发明创造属性的核心技术要素。
# 核心技术要素定义
核心技术要素指的是体现技术创新和进步的关键技术概念，包括但不限于：
- 新兴技术领域和关键技术名称
- 创新的技术方法和系统架构
- 具有进步性的技术特征和解决方案
- 关键的技术模块和核心算法
# 重要注意事项
- 优先从专利名称中识别核心技术创新点
- 结合摘要内容验证和补充技术要素
- 每个要素应为独立的技术词汇或短语（不少于2个字）
- 要素应体现技术的先进性和创造性
- 避免提取应用场景、性能指标、常规部件等非技术核心内容
# 输入格式
专利名称：[专利名称]
专利摘要：[专利摘要]
# 响应说明
- 提供结构化输出，包含提取的所有核心技术要素。严格按照以下JSON格式输出：
{
 "words": ["要素1", "要素2", "要素3", ...]
}
# 相关提取示例
输入：
专利名称：基于露天矿5G网络的数据仿真方法及系统
专利摘要：本发明提供的基于露天矿5G网络的数据仿真方法及系统，涉及人工智能技术领域。在本发明中，获取两个网络拓扑数据；构建出两个网络拓扑数据的第一拓扑数据片段集和第二拓扑数据片段集；基于第一拓扑数据片段集和第二拓扑数据片段集，确定第一个网络拓扑数据的特征向量和第二个网络拓扑数据的特征向量；基于第一个网络拓扑数据的特征向量和第二个网络拓扑数据的特征向量的聚合结果，生成新的网络拓扑数据，以及，在新的网络拓扑数据的仿真结果满足预设仿真需求时，确定仿真结束，并将新的网络拓扑数据作为目标网络拓扑数据。基于上述方法，可以使得数据仿真过程中确定的网络拓扑数据的可靠度可以提高。
输出：
{
 "words": ["5G网络", "数据仿真", "网络拓扑", "特征向量", "人工智能"]
}
输入：
专利名称：基于数字签名的移动支付身份认证方法
专利摘要：本发明涉及支付安全技术领域，具体涉及一种基于数字签名的移动支付身份认证方法。该方法包括：获取设备的追踪信息和用户的行为信息；其中，所述追踪信息用于表征所述设备的身份信息，所述行为信息用于表征所述用户的支付习惯；基于所述追踪信息和所述行为信息，确定交易数字签名；基于所述交易数字签名对移动支付进行验证。本发明公开的方法能够进一步地提高移动支付的安全性。
输出：
{
 "words": ["数字签名", "移动支付", "身份认证"]
}
# 不相关提取示例
- 不提取应用场景："露天矿"
- 不提取常规方法："获取信息"、"验证"
- 不提取性能描述："可靠度"、"安全性"
- 不提取完整句子："生成新的网络拓扑数据"
# 确保你
- 完全按照书写使用标签
- 充分利用专利名称的核心信息
- 每个要素都是独立且有价值的技术概念
- 要素数量控制在1-5个之间，确保质量而非数量
- 不包含任何分析性文字在JSON输出中
"""

def find_items(prompt, text, model=model):
    response = client.chat.completions.create(
        model=model,  # 模型 ID
        messages=[
            {
                'role': 'system',
                'content': prompt
            },
            {
                'role': 'user',
                'content': text
            }
        ]
    )
    return response.choices[0].message.content
```

```python
import json
import pandas as pd

data = pd.read_csv("./中国全量专利数据库1985年.csv")
data = data[['专利名称', '摘要文本']]

temp_dict = {}

for year in range(1985, 1986):
    k = 1
    for i in range(10): # len(data)
        input_text = f"""
        专利名称：{data['专利名称'].iloc[i]}
        专利摘要：{data['摘要文本'].iloc[i]}
        """
        res = json.loads(find_items(prompt, input_text))['words']
        temp_dict[f'{year}_{k}'] = res
        k = k + 1
    
    with open(f"LLM-{year}.json", "w", encoding="utf-8") as f:
        json.dump(temp_dict, f, ensure_ascii=False, indent=4)

    print(f"LLM-{year}.json 导出")
```
